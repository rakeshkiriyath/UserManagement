### Conditional Changes

```python
# For the column Type 1, when value is Fire then change values of column (any column can be given) Type 1 from Fire to  Flamer

df.loc[df['Type 1'] == 'Fire','Type 1'] = 'Flamer'
df
```

```python
# Multiple condition value change
# First take sum up of all columns (HP,Attack,Defense,Sp.Atk,Sp.Def,Speed) as total 
# By default axis = 0 So it will sum up all values in horizontal, We need each count of vertical, So placing axis = 1

'''
+------------+---------+--------+
|            |  A      |  B     |
+------------+---------+---------
|      0     | 0.626386| 1.52325|----axis=1----->
+------------+---------+--------+
             |         |
             | axis=0  |
             ↓         ↓
'''

df['Total'] = df.loc[:,'HP':'Speed'].sum(axis=1)
df.loc[df['Total'] >= 500 ,['Legendary','Generation']] = [True,7]
```

```python
df.head()
```

```python
df.drop('Total',axis=1,inplace=True)
df.head()
```

### Aggretate Statistics

```python
# Group By : Choose a specific column and group with other columns to find any useful informations

'''
    mean()
    sum()
    count()
'''

df.groupby(['Type 1','Type 2']).mean()[['Defense','Attack','Speed']].sort_values(by=['Defense','Attack','Speed'],ascending=False)
```
